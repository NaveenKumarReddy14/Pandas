**Pandas Series**
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.


```python
a = [10,23,45,'naveen','reddy',21]
x= pd.Series(a)
print(x)
```

```python
x[0],x[3] # return the values based on the index position
```

```python
x=pd.Series(a,index=['a','b','c','d','e','f']) #When you have created labels, you can access an item by referring to the label.
print(x)
x['a'],x['d']
```

```python
info = {'name' : 'naveen', 'age': 21, 'state': 'Ap'} #use a key/value object, like a dictionary, when creating a Series.
x=pd.Series(info)
print(x)
```

```python
x=pd.Series(info,index=['name','state'])
print(x)
```

**DataFrames**
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

```python
std={
    'name':['naveen','kumar','hethwik',],  #Create a DataFrame from three Series:
    'Age' :[21,20,11],
    'country':['UK','USA','UK']
}
df=pd.DataFrame(std)
print(df)
```

```python
df.loc[0] #Return row 0:
```

```python
df.loc[[0,1]] #use a list of indexes:
```
